Create 2 arrays to make a plot.

```python
x = np.arange(0.,10.,0.1)
y = x

# plt.plot(x,y)
plt.plot(x,y, x, y**2)
# plt.show()
```

```python
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Simple Plot')
plt.legend()
```

**Linewidth**

```python
plt.plot(x, np.sin(x), x, np.cos(x), linewidth=5)
plt.legend(['sin(x)', 'cos(x)'])
```

<a id=colors></a>
**Colors**

https://matplotlib.org/stable/gallery/color/named_colors.html

```python
plt.plot(x, np.sin(x), 'k', x, np.cos(x), 'b', linewidth=5)
plt.legend(['sin(x)', 'cos(x)'])
```

<a id=markers></a>
**Markers**

https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(x, np.sin(x), 'r<', x, np.cos(x), 'g>', markersize=8)
plt.legend(['sin(x)', 'cos(x)'])
```

<a id=subplot></a>
**Subplot**

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

```python
# subplot(n_rows, n_cols, n_plots)
plt.subplot(121)
plt.plot(x, np.exp(x))

plt.subplot(122)
plt.plot(x, x**3)
```

```python
fig = plt.figure(figsize=(10,8))

plt.subplot(211)
plt.plot(x, np.sin(x), label='sin(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Plot 1')
plt.legend()


# 2nd plot
plt.subplot(212)

# use "linestyle" to specify linestyle:  linestyle='--'
# and "marker" for markers:  marker='d'
plt.plot(x, np.cos(x), color='r', label='cos(x)')

plt.xlabel('x2')
plt.ylabel('y2')

plt.title('Plot 2')
plt.legend()
plt.tight_layout()

# use dpi (dots per inch) to have a bigger resolution. The default dpi is 100.
fig.savefig('fig.png')
```
